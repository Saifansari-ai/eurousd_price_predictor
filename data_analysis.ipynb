# data analysis on Euro_usd_historical_data
## Data cleaning


```python
# null values in the dataset
df.isnull().sum()
```

* the Vol. column has no values 
* so romove the column Vol.

```python
# drop Vol. column
df.drop('Vol.', axis=1, inplace=True)
df.head(-1)
```

```python
# romove unnamed column
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head(-1)
```

```python
# null values in the dataset
df.isnull().sum()
```

* now there is no null values in the dataset

```python
# info of the dataset
df.info()
```

* here the change% is object datatype but we need it as a float in order to do analysis
* so changing the change % to float
* the Date column is stored as string so change to Date

```python
# changing the Change % to percentage between 0 and 1
df['Change %'] = df['Change %'].str.rstrip('%').astype('float') / 100
```

```python
# change Date string to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
```

```python
df.head()
```

* now the values of change% is in between 0 and 1 to analyse better

```python
# discription of the dataset
df.describe()
```

* no duplicates values in the dataset
* no null values in the dataset


```python
df.corr()
```

# Visualization on Euro-Usd dataset

```python
# histogram on price,open,high,low
plt.figure(figsize=(20,20))
df.hist(column=['Price','Open','High','Low'])
plt.tight_layout()
plt.show()
```

* the distribution of price,close,high and low are symmitical
* there is high correlation between these columns

```python
# time vs price
plt.figure(figsize=(15,8))
plt.plot(df['Date'],df['Price'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Time vs Price')
plt.show()
```

* price from 2003 to 2008 were booming
* then 2008 crisis happend then there were fall back of the market 
* after that the price has never recovered 
* from 2008 to 2015 the price were on avrg 1.3 
* then in 2015 the prices suddenly drop from 1.4 to 1.0 after that it never recover
* after 2015 the prices stay on avrg 1.0
* so overall from 2008 onwards the price is falling


```python
# Heatmap of Correlations 
plt.figure(figsize=(15,8))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

* price,open,high,low has perfect score of 1 means there are completely correlated
