# Feature engineering on euro-usd dataset

### moving average

```python
df.sort_values(by="Date", inplace=True)

# Calculate Simple Moving Averages (SMA)
df["SMA_7"] = df["Price"].rolling(window=7, min_periods=1).mean()
df["SMA_30"] = df["Price"].rolling(window=30, min_periods=1).mean()
df["SMA_100"] = df["Price"].rolling(window=100, min_periods=1).mean()
df["SMA_200"] = df["Price"].rolling(window=200, min_periods=1).mean()

# Save the updated dataset (overwrite original CSV)
df.to_csv("EUR_USD_price_2000_2025.csv", index=False)

```

```python
df.head()
```

### Day of the week for every row

```python
# Day of the week with name
df['Day'] = df['Date'].dt.day_name()
df.to_csv("EUR_USD_price_2000_2025.csv", index=False)
df.sample(5)
```

### week of the year

```python
# week of the year
df["Week"] = df["Date"].dt.isocalendar().week
df.to_csv("EUR_USD_price_2000_2025.csv", index=False)
df.sample(5)
```

### Month of the year

```python
# month of the year
df["Month"] = df["Date"].dt.month
df.to_csv("EUR_USD_price_2000_2025.csv", index=False)
df.sample(5)
```

```python
# year of the date
df["Year"] = df["Date"].dt.year
df.to_csv("EUR_USD_price_2000_2025.csv", index=False)
df.sample(5)
```

### RSI of the data

```python
# Calculate daily price change
df["Price_Change"] = df["Price"].diff()

# Separate gains and losses
df["Gain"] = df["Price_Change"].apply(lambda x: x if x > 0 else 0)
df["Loss"] = df["Price_Change"].apply(lambda x: -x if x < 0 else 0)

# Calculate the 14-day average gain and loss
df["Avg_Gain"] = df["Gain"].rolling(window=14, min_periods=1).mean()
df["Avg_Loss"] = df["Loss"].rolling(window=14, min_periods=1).mean()

# Compute the Relative Strength (RS)
df["RS"] = df["Avg_Gain"] / df["Avg_Loss"]

# Compute RSI
df["RSI_14"] = 100 - (100 / (1 + df["RS"]))

# Save the updated dataset
df.to_csv("EUR_USD_price_2000_2025.csv", index=False)

df.sample(5)
```

### Moving Average Convergence Divergence

```python
# Calculate the 12-day and 26-day EMA
df["EMA_12"] = df["Price"].ewm(span=12, adjust=False).mean()
df["EMA_26"] = df["Price"].ewm(span=26, adjust=False).mean()

# Calculate MACD Line
df["MACD"] = df["EMA_12"] - df["EMA_26"]

# Calculate Signal Line (9-day EMA of MACD)
df["Signal_Line"] = df["MACD"].ewm(span=9, adjust=False).mean()

# Save the updated dataset
df.to_csv("EUR_USD_price_2000_2025.csv", index=False)

df.sample(5)
```

### lag feature of 1,3,7 days
**1-Day Lag (Lag_1)**
* Captures short-term trends
* Helps predict immediate price movement
* Useful for high-frequency trading models

**3-Day Lag (Lag_3)**
* Identifies weekly cycle patterns
* Good for short-term swing trading
* Helps detect recent momentum shifts

**7-Day Lag (Lag_7)**
* Captures weekly trends
* Useful for medium-term trading strategies
* Can detect patterns in forex market cycles

```python
# Add lag features
df["Lag_1"] = df["Price"].shift(1)
df["Lag_3"] = df["Price"].shift(3)
df["Lag_7"] = df["Price"].shift(7)


# Save the updated dataset
df.to_csv("EUR_USD_price_2000_2025.csv", index=False)
df.head(10)
```

### Creating a target price column from price column
* Target_Price that stores the next day's price, making it the target variable for a forecasting model

```python
df['Target_Price'] = df['Price'].shift(-1)
df.to_csv("EUR_USD_price_2000_2025.csv", index=False)
df.head(10)
```

### Adding lag_14 and lag_30 to capture longer trends

```python
df["Lag_14"] = df["Price"].shift(14)
df["Lag_30"] = df["Price"].shift(30)

# Save the updated dataset
df.to_csv("EUR_USD_price_2000_2025.csv", index=False)
df.head()
```

## Implementing volatility metrics to capture volatility
1. Bollinger Bands
2. ATR

### Bollinger Bands

```python
# Calculate 20-day Simple Moving Average (SMA)
df['SMA_20'] = df['Price'].rolling(window=20).mean()

# Calculate Standard Deviation
df['STD_20'] = df['Price'].rolling(window=20).std()

# Calculate Upper and Lower Bollinger Bands
df['Upper_Band'] = df['SMA_20'] + (2 * df['STD_20'])
df['Lower_Band'] = df['SMA_20'] - (2 * df['STD_20'])

# Drop NaN values (first 20 rows will be NaN due to rolling window)
df = df.dropna()

# Save the updated dataset
df.to_csv("EUR_USD_price_2000_2025.csv", index=False)
df.head()
```

### ATR

```python

# ATR Calculation
df['High-Low'] = df['High'] - df['Low']
df['High-Close'] = abs(df['High'] - df['Price'].shift(1))
df['Low-Close'] = abs(df['Low'] - df['Price'].shift(1))

# True Range (TR) is the max of the above three
df['TR'] = df[['High-Low', 'High-Close', 'Low-Close']].max(axis=1)

# ATR (14-day moving average of TR)
df['ATR_14'] = df['TR'].rolling(window=14).mean()

# Drop NaN values (first 14 rows will be NaN due to rolling window)
df = df.dropna()

# Save the updated dataset
df.to_csv("EUR_USD_price_2000_2025.csv", index=False)
df.head()
```

# Visualization on feature engineered columns

### Visualization on the simple moving average
**SMA_7 vs SMA_30**

```python
# Define 5-year phases
years = list(range(2000, 2025, 5))
num_phases = len(years)

# Determine subplot grid size
rows = (num_phases // 2) + (num_phases % 2)  # Ensures enough rows
cols = 2  # 2 columns per row

# Create figure with multiple subplots
fig, axs = plt.subplots(rows, cols, figsize=(16, rows * 4))  
axs = axs.flatten()  # Flatten for easy indexing

# Plot each 5-year phase
for i, start_year in enumerate(years):
    # Filter data for the 5-year range
    df_subset = df[(df['Date'].dt.year >= start_year) & (df['Date'].dt.year < start_year + 5)]
    
    # Plot SMA_7 and SMA_30
    axs[i].plot(df_subset['Date'], df_subset['SMA_7'], label='SMA 7', color='blue')
    axs[i].plot(df_subset['Date'], df_subset['SMA_30'], label='SMA 30', color='orange')

    # Set labels and title
    axs[i].set_title(f'{start_year}-{start_year+4} (SMA_7 vs SMA_30)')
    axs[i].set_xlabel('Date')
    axs[i].set_ylabel('SMA Value')
    axs[i].legend()
    axs[i].grid(True)

# Remove extra empty subplots
for j in range(i + 1, len(axs)):
    fig.delaxes(axs[j])

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

## Conclusion

**When SMA_7 crosses above SMA_30 ("Golden Cross")**
- **📈 Bullish Signal** – This suggests a potential uptrend, indicating that prices may continue rising. Traders may consider buying or holding long positions.

**When SMA_7 crosses below SMA_30 ("Death Cross")**
- **📉 Bearish Signal** – This suggests a potential downtrend, indicating that prices may continue falling. Traders may consider selling or shorting the asset.

**If SMA_7 stays above SMA_30 for an extended period**
- **🔼 Strong Uptrend** – Momentum is positive, and prices are likely in a sustained bullish phase.

**If SMA_7 stays below SMA_30 for an extended period**
- **🔽 Strong Downtrend** – Momentum is negative, and prices may keep declining.

## Visualization on the Temporal feature of the data

**day**

```python
# Compute volatility (standard deviation of returns) for each weekday
weekday_volatility = df.groupby('Day')['Price_Change'].std()

# Reorder weekdays for better visualization
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
weekday_volatility = weekday_volatility.reindex(weekday_order)

# Plot the volatility for each weekday
plt.figure(figsize=(10, 5))
sns.barplot(x=weekday_volatility.index, y=weekday_volatility.values, palette='coolwarm')
plt.title('Volatility (Standard Deviation of Returns) by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Volatility (Std Dev of Returns)')
plt.grid(axis='y')

plt.show()
```

### Conclusion
- the forex market on only open on weekdays
- the market is less volatile on the opening day (monday)
- throughout the week the volatility increases 
- on the closing day of the weekday or of market(friday) the maket is most volatile due to closing of the market
- then market open on monday

### week

```python
# Compute weekly volatility (standard deviation of returns for each week)
weekly_volatility = df.groupby('Week')['Price_Change'].std()

# Plot the volatility for each week of the year
plt.figure(figsize=(12, 6))
sns.lineplot(x=weekly_volatility.index, y=weekly_volatility.values, marker='o', color='blue')
plt.title('Volatility (Std Dev of Returns) by Week of the Year')
plt.xlabel('Week of the Year')
plt.ylabel('Volatility (Std Dev of Returns)')
plt.grid(True)

plt.show()
```

## Conclusion
**High Volatility Periods:**

- The highest volatility occurs around Week 10 (early March).
- Another increase is seen between Weeks 40-45 (October-November).
- These spikes might be linked to economic announcements, geopolitical events, or seasonal trends.

**Low Volatility Periods:**

- Around Week 50-52 (December), volatility drops sharply, likely due to reduced trading activity near the year-end holidays.
- There are also brief dips around Weeks 15-20 (April-May), which may indicate calmer market conditions.

**General Trend:**

- Volatility fluctuates throughout the year but generally remains between 0.005 and 0.008.
- Sudden spikes suggest market uncertainty, while stable periods indicate lower price fluctuations.

## Months

```python
# Compute volatility (standard deviation of returns) per month
monthly_volatility = df.groupby('Month')['Price_Change'].std()

# Plot the monthly volatility
plt.figure(figsize=(10, 5))
plt.plot(monthly_volatility.index, monthly_volatility, marker='o', linestyle='-', color='blue')
plt.xlabel('Month')
plt.ylabel('Volatility (Std Dev of Returns)')
plt.title('Volatility by Month')
plt.grid(True)
plt.show()
```

## Conclusion
**High Volatility Months:**

- March (3rd month) and April (4th month) show significant volatility spikes.
- October (10th month) and November (11th month) also exhibit increased fluctuations.
- This could be due to economic reports, central bank decisions, or major financial events occurring in these months.

**Low Volatility Months:**

- February (2nd month), May (5th month), and August (8th month) show lower volatility.
- This could be due to reduced trading activity, seasonal market effects, or lower economic uncertainties.

**Trading Implications:**

- Higher volatility months indicate greater price fluctuations, providing more trading opportunities but also increased risk.
- Lower volatility months suggest stable trends, which may be better suited for conservative strategies like long-term holding.

**Potential Actions:**

- If trading short-term, focus on high-volatility months for potential breakout trades.
- If investing long-term, low-volatility months may offer better entry points for stable trends.
